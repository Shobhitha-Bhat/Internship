BANK CUSTOMER CHURN CLASSIFICATION MODEL


OBJECTIVE:
Its a model that helps financial institutions to predict customers who are likely to leave(churn) in near future.
It makes use of various customer details like their:
->Credit score
->Geography
->Age
->Tenure
->Balance
->Number of products
->Estimated salary
->Activeness
->Posession of a credit card

DATA SOURCE: https://github.com/YBI-Foundation/Dataset
            ->https://github.com/YBI-Foundation/Dataset/blob/main/Bank%20Churn%20Modelling.csv

```python
#IMPORTING LIBRARY
import pandas as pd
```

```python
#IMPORTING DATA
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Bank%20Churn%20Modelling.csv')
```

DEFINING TARGET VARIABLE(y) AND FEATURE VARIABLES(x)

```python
y = df['Churn']
```

```python
x = df[['CustomerId', 'CreditScore', 'Geography', 'Gender', 'Age',
       'Tenure', 'Balance', 'Num Of Products', 'Has Credit Card',
       'Is Active Member', 'Estimated Salary']]
```

```python
x
```

ENCODING

```python
df['Geography'].value_counts()
```

```python
df.replace({'Geography':{'France':0,'Germany':1,'Spain':2}},inplace=True)
```

```python
df['Surname'].value_counts
```

```python
df['Gender'].value_counts()
```

```python
df.replace({'Gender':{'Male':0,'Female':1}},inplace=True)
```

```python
x = df[['CustomerId', 'CreditScore', 'Geography', 'Gender', 'Age',
       'Tenure', 'Balance', 'Num Of Products', 'Has Credit Card',
       'Is Active Member', 'Estimated Salary']]
```

```python
#TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#MODELING
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
#MODEL EVALUATION
model.fit(x_train,y_train)
```

```python
#MODEL PREDICTION
ypred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
#ACCURACY SCORES
print(classification_report(y_test,ypred))
```
